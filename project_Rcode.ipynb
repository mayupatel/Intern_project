#### Shapiro Test for check normality of the Data:

- Conducting Shapiro's test for original dataset:

```r
#shapiro test for whole dataset named cc_metabo:
#Here, the p-values are generated for each and every metabolite IDs.


#assigning to store the values in here.
metaboData <- c()

#running 'for' loop for each and every column that contain numeric values.
for (i in 3:ncol(cc_metabo)){
    
  #conducting the test and assining it to ST
  ST = shapiro.test(as.numeric(cc_metabo[[i]]))
  #from ST picking p-values and storing it in local variable and concatenate them
  metaboData <- c(metaboData,ST$p.value)
      
}

#printing the p-values:
print("The p-values generated from the Shapiro-Wilk test for metabolite IDs")
print(metaboData) # p-values of numeric data is shown
```

- Looking for p-values < 0.05

```r
#assigning the variable to store the p-values
pVal <- c()

#looping the metabodata and storing p-values <0.05
for (m in (metaboData)){
    
    #if statement to look for values < 0.05
    if (m < 0.05){
        #concatenate the values to local variable pval
        pVal <- c(pVal,m)
    }   
}
#printing the p-values
pVal
```

- As we can see, only few of the p-values are less than 0.05, suggesting most to be greater than 0.05. Thus, the given data is normally distributed.

### t-test for both subset data:

#First thing to proceed with t-test is the given dataset contains n <30 sample size. Here, the t-test is calculated between each treatment group and control group to produce p-values.

```r
# Writing a function which takes subset data as an argument to produce resultant output.

tTest_calculation <- function(subset) {
    
    #Store column control group into a variable and calculate mean and SD.
    control = subset[,8]
    #Mean of control group
    mean_control <- mean(control)
    #Standard deviation of control group
    sd_control <- sd(control)
    
    #Assigning variables to store treatment group mean and standard deviation 
    mean_treat <- c()
    sd_treat <- c()
    
    #Assigning variable to store generated pvalues from resultant t-test
    raw_pval <- c()
    
    #Using column names from subset data excluding control group are stored in a variable
    treatment_names <- colnames(subset[,-8])
    
    #For loop for every column of subset data except control group, so adding -1.
    for (i in 1:(ncol(subset)-1)){
        
        #Generating mean and sd values for every column and stored into the assigned variable.
        mean_treat <- c(mean_treat,mean(subset[,i]))
        sd_treat <- c(sd_treat,sd(subset[,i]))
        
        #Using variance test to proceed with t-test.
        #calculating variance test between treatment and control group to generate p-value
        var_result =var.test(subset[,i],control,data=subset)
        var_pval = var_result$p.value
        
        # Using the resultant p-value from variance test for if/else statement
        if (var_pval < 0.05){
            # p-value is less than 0.05, then welch t.test is performed as variance are not equal
            #The resultant output is stored as raw p-value
            result <- t.test(subset[,i], control,data=subset,var.equal=FALSE)
            raw_pval <- c(raw_pval,result$p.value) 
    
        } else {
            #p-value is greater than 0.05, then two-sample t.test is performed as variance are equal
            #The resultant output is stored as raw p-value
            result <- t.test(subset[,i], control,data=subset,var.equal=TRUE,paired=FALSE) 
            raw_pval <- c(raw_pval,result$p.value)
    
        }
    }
   #printing the mean and sd of control group 
   print(paste("The Mean value of Control group is:",mean_control))
   print(paste("The Standard Deviation of Control group is:",sd_control))
    
   #for treatment groups
   #Binding the variables to produce table like format as an output of treatment groups.
   cbind(treatment_names,raw_pval,mean_treat,sd_treat)
       
}
#Exit the function


print("For 30minutes subset data between treatment and control groups::")
# passing 30min subset data to a function.
tTest_calculation(subset_30min)

print("For 90minutes subset data between treatment and control groups::")  
# passing 90min subset data to a function.
tTest_calculation(subset_90min)
```

The resultant output of p-values shows that they are > 0.05, suggesting that there is no significant differences between treatment and control group.

### t-test for fold change values between 30 and 90 minutes subset data:

#First thing to proceed with t-test is the given fold change dataset contains n<30 sample size. Here, the t-test is calculated between 30 and 90 mins subset data for every treatment group containing fold change values.

```r
#Writing a function and passing two arguments of both subset data into a function.

tTest_FoldChange <- function(FC_30min,FC_90min) {
    
    #Storing the mean and sd of 30 mins treatment groups
    mean_FC30min <- c()
    sd_FC30min <- c()
    
    #Storing the mean and sd of 90 mins treatment groups
    mean_FC90min <- c()
    sd_FC90min <- c()
    
    #Assigning variable to store generated pvalues from t-test
    rawPval_FC <- c()
    
    #Using column names from subset data for all the treatment groups
    treatment_names <- colnames(FC_30min[,2:8])
    
    #For loop used for both dataset by taking each column from either of them and performing t-test
    for (i in 2:ncol(FC_30min)){
        for (j in 2:ncol(FC_90min)){    
        }
        
        #Generating mean and sd values for 30 mins treatment groups
        mean_FC30min <- c(mean_FC30min,mean(FC_30min[,i]))
        sd_FC30min <- c(sd_FC30min,sd(FC_30min[,i]))
        
        #Generating mean and sd values for 90 mins treatment groups
        mean_FC90min <- c(mean_FC90min,mean(FC_90min[,j]))
        sd_FC90min <- c(sd_FC90min,sd(FC_90min[,j]))
        
        #Using variance test before performing t-test.
        #Calculating variance test between 30- and 90-mins data to generate p-value
        var_FC = var.test(FC_30min[,i],FC_90min[,j])
        varPval_FC = var_FC$p.value
        
        # Using the resultant p-value from variance test for if/else statement
        if (varPval_FC < 0.05){
            # p-value is less than 0.05, then welch t.test is performed as variance are not equal
            #The resultant output is stored as raw p-value
            result <- t.test(FC_30min[,i], FC_90min[,j],var.equal=FALSE)
            rawPval_FC <- c(rawPval_FC,result$p.value) 
    
        } else {
            #p-value is greater than 0.05, then two-sample t.test is performed as variance are equal
            #The resultant output is stored as raw p-value
            result <- t.test(FC_30min[,i],FC_90min[,j],var.equal=TRUE,paired=FALSE) 
            rawPval_FC <- c(rawPval_FC,result$p.value)
        }
                   
    }
   #Binding the variables to produce table lie format to view resultant output of treatment groups.
   cbind(treatment_names,rawPval_FC,mean_FC30min,sd_FC30min,mean_FC90min,sd_FC90min)
}


print("Comparing 30- and 90-mins treatment groups based on fold change values:")
#passing parameters of subset data to conduct t-test.
tTest_FoldChange(FC_30min,FC_90min)
    
```

The resultant output of p-values shows that they are few treatment groups whose pvalues are < 0.05, suggesting that there is significant differences between those treatment groups.
The list of those groups are:
1. 15mM Glucose 
2. 15mM (0.65 fructose, 0.35 Glucose)
3. 15mM (0.85 fructose, 0.15 Glucose)
4. 15mM fructose

### ANOVA test for both subset data:

#First thing to proceed with ANOVA test is the given dataset contains n <30 sample size. Here, ANOVA is calculated between treatment groups to produce p-values.

```r
# Writing a function which takes subset data as an argument to produce resultant output for ANOVA

Anova_subsetDt <- function(subsetData){
    
    #Here, the subset data is transposed and converted to dataframe for performing ANOVA
    combined_groups <- data.frame(t(subsetData[3:70]))
    #View the summary of the combine groups
    summary(combined_groups)
    
    #These groups are stacked together for further analysis
    stack_groups <- stack(combined_groups)
    #The stack group columns are re-named as shown below
    colnames(stack_groups) <- c("metabolite_levels","treatment")
    
    #print(bartlett.test(stack_groups$metabolite_levels,stack_groups$treatment))
    
    #Running ANOVA test on the stack group columns
    anova_Result <- aov(metabolite_levels~treatment,data=stack_groups)
    #Printing summary of the anova
    print(summary(anova_Result))
    
    #This helps in extracting the p-value from the anova summary result
    anov_pval <- summary(anova_Result)[[1]][["Pr(>F)"]][[1]]
    
    #Here, conditional statement is important because if p-value is < 0.05,
    #then there is significant difference in the resultant group to conduct a Tukey test
    if (anov_pval < 0.05){
        #Conduct a Tukey's multiple comparsion procedure to know the difference in the groups
        #Taking anova result on which Tukey test is performed
        tukey_test <- TukeyHSD(aov(metabolite_levels~treatment,data=stack_groups))
        print(tukey_test)
        
        #Plotting the tukey test result
        plot(tukey_test)
        
    }        
    
}
#Exit the function

print("ANOVA test for 30-and 90-mins subset data")

#Passing 30mins subset data to a function.
print("FOR 30Mins SUBSET DATA")
Anova_subsetDt(data30Min)

#Passing 90mins subset data to a function.
print("FOR 90Mins SUBSET DATA")
Anova_subsetDt(data90Min)
```

p-value is > 0.05, suggesting that there is no significant difference between the mean values.

Based on the p-adj values, we can say that none of them are statistically significant as the p-values are greater than 0.05.

### ANOVA test for Fold change values of both subset data:

#First thing to proceed with ANOVA test is the given dataset contains n <30 sample size. Here, ANOVA is calculated between treatment groups to produce p-values.

```r
# Writing a function which takes Fold change values data as an argument to produce resultant output for ANOVA

Anova_FoldChangDt <- function(FC_data){
    
    #Here, the subset data is transposed and converted to dataframe for performing ANOVA
    FC_groups <- data.frame(FC_data[,2:8])
    #View the summary of the combine groups
    summary(FC_groups)
    
    #These groups are stacked together for further analysis
    stack_FC <- stack(FC_groups)
    #The stack group columns are re-named as shown below
    colnames(stack_FC) <- c("metabolite_levels","treatment")
    
    #print(bartlett.test(stack_FC$metabolite_levels,stack_FC$treatment))
    
    #Running ANOVA test on the stack group columns
    anova_FC <- aov(metabolite_levels~treatment,data=stack_FC)
    #Printing summary of the anova
    print("ANOVA TEST")
    print(summary(anova_FC))
    
    #This helps in extracting the p-value from the anova summary result
    anovFC_pval <- summary(anova_FC)[[1]][["Pr(>F)"]][[1]]
    
    #Here, conditional statement is important because if p-value is < 0.05,
    #then there is significant difference in the resultant group to conduct a Tukey test
    if (anovFC_pval < 0.05){
        #Conduct a Tukey's multiple comparsion procedure to know the difference in the groups
        #Taking anova result on which Tukey test is performed
        tukey_FC <- TukeyHSD(aov(metabolite_levels~treatment,data=stack_FC))
        print("TUKEY TEST")
        print(tukey_FC)
        
        #Plotting the tukey test result.
        plot(tukey_FC)
   } 
}
#Exit the function

print("ANOVA test for 30-and 90-mins Fold change value subset data")
```

```r
#Passing 30mins subset data to a function.
print("FOR 30Mins FC SUBSET DATA")
Anova_FoldChangDt(FC_30min)
```

```r
#Passing 90mins subset data to a function.
print("FOR 90Mins FC SUBSET DATA")
Anova_FoldChangDt(FC_90min)
```

p-value < 0.05, showing statistical significance with the difference in the mean values of the different treatment groups.
