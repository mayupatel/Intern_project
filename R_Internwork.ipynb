```r
#p-values are generated for each metabolite
#shapiro test for the whole dataset:
metaboData <- c()
for (i in 3:ncol(cc_metabo)){
  ST = shapiro.test(as.numeric(cc_metabo[[i]]))
  metaboData <- c(metaboData,ST$p.value)
      
}
print("The p-values generated from the Shapiro-Wilk test")
print(metaboData) # p-values of numeric data is shown
hist(metaboData)  # generate histogram to show frequency of each p-values
```

## t-test for each subset data:

```r
# writing a function which each time takes a subset data to produce the output.

tTest_calculation <- function(subset) {
    
    # store column control into a variable and calculate mean and SD.
    control = subset[,8]
    mean_control <- mean(control)
    sd_control <- sd(control)
    
    #storing treatment group mean and sd.
    mean_treat <- c()
    sd_treat <- c()
    
    #assigning variable to store generated pvalues from t-test
    raw_pval <- c()
    
    # using column names from subset data excluding control group
    treatment_names <- colnames(subset[,-8])
    
    # loop for columns of subset data
    for (i in 1:(ncol(subset)-1)){
        
        #generating mean and sd values.
        mean_treat <- c(mean_treat,mean(subset[,i]))
        sd_treat <- c(sd_treat,sd(subset[,i]))
        
        # using variance test to conduct t-test.
        var_result =var.test(subset[,i],control,data=subset)
        var_pval = var_result$p.value
        
        # using the p-values from variance test
        if (var_pval < 0.05){
            # p-value is less than 0.05, then welch t.test is performed as variance are not equal
            result <- t.test(subset[,i], control,data=subset,var.equal=FALSE)
            raw_pval <- c(raw_pval,result$p.value) 
    
        } else {
            #p-value is greater than 0.05, then two-sample t.test is performed as variance are equal
            result <- t.test(subset[,i], control,data=subset,var.equal=TRUE,paired=FALSE) 
            raw_pval <- c(raw_pval,result$p.value)
    
        }
    }
    #printing the mean and sd of control group 
   print(paste("The mean value of Control group is:",mean_control))
   print(paste("The Standard Deviation of Control group is:",sd_control))
    
   # binding the variables to produce table format output of treatment groups.
   cbind(treatment_names,raw_pval,mean_treat,sd_treat)
       
}

# passing subset data to a function every time.
print("For 30minutes subset data between treatment and control groups::")
tTest_calculation(subset_30min)
print("For 90minutes subset data between treatment and control groups::")  
tTest_calculation(subset_90min)
```

The resultant output of p-values shows that they are > 0.05, suggesting that there is no significant difference between treatment and control group.

## t-test between 30 and 90 minutes subset data of fold change values at each concentration:

```r
# writing a function for fold change data files
tTest_FoldChange <- function(FC_30min,FC_90min) {
    
    #storing the mean and sd of 30 mins treatment groups
    mean_FC30min <- c()
    sd_FC30min <- c()
    #storing the mean and sd of 90 mins treatment groups
    mean_FC90min <- c()
    sd_FC90min <- c()
    
    #assigning variable to store generated pvalues from t-test
    rawPval_FC <- c()
    
    # using column names from subset data excluding control group
    treatment_names <- colnames(FC_30min[,2:8])
    
    # for loop used for both dataset
    for (i in 2:ncol(FC_30min)){
        for (j in 2:ncol(FC_90min)){    
        }
        #generating mean and sd values for 30 mins treatment groups
        mean_FC30min <- c(mean_FC30min,mean(FC_30min[,i]))
        sd_FC30min <- c(sd_FC30min,sd(FC_30min[,i]))
        
        #generating mean and sd values for 90 mins treatment groups
        mean_FC90min <- c(mean_FC90min,mean(FC_90min[,j]))
        sd_FC90min <- c(sd_FC90min,sd(FC_90min[,j]))
        
        # using variance test to conduct t-test.
        var_FC = var.test(FC_30min[,i],FC_90min[,j])
        varPval_FC = var_FC$p.value
        # using the p-values from variance test
        if (varPval_FC < 0.05){
             # p-value is less than 0.05, then welch t.test is performed as variance are not equal
            result <- t.test(FC_30min[,i], FC_90min[,j],var.equal=FALSE)
            rawPval_FC <- c(rawPval_FC,result$p.value) 
    
        } else {
            #p-value is greater than 0.05, then two-sample t.test is performed as variance are equal
            result <- t.test(FC_30min[,i],FC_90min[,j],var.equal=TRUE,paired=FALSE) 
            rawPval_FC <- c(rawPval_FC,result$p.value)
    
        }
                   
    }
    # binding the variables to produce table format output of treatment groups.
   cbind(treatment_names,rawPval_FC,mean_FC30min,sd_FC30min,mean_FC90min,sd_FC90min)
}

print("Comparing 30- and 90-mins treatment groups based on fold change values:")
#passing two files to conduct t-test.
tTest_FoldChange(FC_30min,FC_90min)
    
```

The resultant output of p-values shows that they are few treatment groups whose p-values are < 0.05, suggesting that there is significant difference between those treatment groups. The list of those groups are:

1. 15mM Glucose
2. 15mM (0.65 fructose, 0.35 Glucose)
3. 15mM (0.85 fructose, 0.15 Glucose)
4. 15mM fructose

##  ANOVA for each subset data:

```r
Anova_subsetDt <- function(subsetData){
    
    
    combined_groups <- data.frame(t(subsetData[3:70]))
    summary(combined_groups)
    stack_groups <- stack(combined_groups)
    colnames(stack_groups) <- c("metabolite_levels","treatment")
    #print(bartlett.test(stack_groups$metabolite_levels,stack_groups$treatment))
    anova_Result <- aov(metabolite_levels~treatment,data=stack_groups)
    print(summary(anova_Result))
    anov_pval <- summary(anova_Result)[[1]][["Pr(>F)"]][[1]]
    #here conditional statement is important because if p-value is < 0.05,
    #then there is significant difference to conduct Tukey test
    if (anov_pval < 0.05){
        #conduct a Tukey's multiple comparsion procedure to know the difference in the groups
        tukey_test <- TukeyHSD(aov(metabolite_levels~treatment,data=stack_groups))
        print(tukey_test)
        plot(tukey_test)
        
    }        
    
}
print("ANOVA test for 30-and 90-min subset data")
Anova_subsetDt(data30Min)

Anova_subsetDt(data90Min)
```

p-value is > 0.05, suggesting that there is no significant difference between the mean values.

## ANOVA for the foldchange data:

```r
Anova_FoldChangDt <- function(FC_data){
    
    
    FC_groups <- data.frame(FC_data[,2:8])
    summary(FC_groups)
    stack_FC <- stack(FC_groups)
    colnames(stack_FC) <- c("metabolite_levels","treatment")
    #print(bartlett.test(stack_FC$metabolite_levels,stack_FC$treatment))
    anova_FC <- aov(metabolite_levels~treatment,data=stack_FC)
    print(summary(anova_FC))
    anovFC_pval <- summary(anova_FC)[[1]][["Pr(>F)"]][[1]]
    if (anovFC_pval < 0.05){
        tukey_FC <- TukeyHSD(aov(metabolite_levels~treatment,data=stack_FC))
        print(tukey_FC)
        plot(tukey_FC)
   } 
}

Anova_FoldChangDt(FC_30min)

Anova_FoldChangDt(FC_90min)
```

p-value < 0.05, showing statistical significance with the difference in the mean values of the different treatment groups

11 groups show the differences in the mean values
